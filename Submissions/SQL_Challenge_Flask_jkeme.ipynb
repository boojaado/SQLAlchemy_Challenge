```python
# Import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sqlalchemy import create_engine, inspect
```

```python
# Week 10, Day 3, Activity 10 - Flask_with_ORM
path = "sqlite:///Resources/hawaii.sqlite"
engine = create_engine(path)

# Establish connection
conn = engine.connect()
```

```python
# Week 10, Day 3, Activity 10 - Flask_with_ORM
inspector = inspect(engine)

tables = inspector.get_table_names()

tables
```

```python
# First query to be inserted into the flask app.
# Create a dictionary using date as the key and prcp as the value.
# Return a JSON representation of dictionary using date as the key and prcp as the value.

query1 = """
        SELECT
            date,
            station,
            prcp
        FROM
            measurement
        ORDER BY
            date asc,
            station asc
        """

precipitation_df = pd.read_sql(query1, conn)

precipitation_df
```

```python
# Second query to be inserted into the flask app.
# Create a dictionary of station.
# Return a JSON list of stations from the dataset.

query2 = """
        SELECT
            station,
            name
        FROM
            station;
        """
station_df = pd.read_sql(query2, conn)
station_df
```

```python
# Third query to be inserted into the flask app.
# Create a dates and temperature observations of the most active station for the last year of data.
# Return a JSON list of temperature observations (TOBS) for the previous year.

query3 = """
    SELECT
      station,
      date,
      tobs

    FROM
        measurement
    WHERE
        date >= '2016-08-23'
    AND
        station = 'USC00519281'
    ORDER BY
        date asc;
    """
tobs_df = pd.read_sql(query3, conn)
tobs_df
```

```python
# Fourth query to be inserted in the flask app.
# Return a JSON list of the minimum temperature, the average temperature, 
# and the max temperature for a given start-date.

start_date = "2017-08-17"

query4 = f"""
    SELECT
        min(tobs) as TMIN,
        avg(tobs) as TAVG,
        max(tobs) as TMAX
    FROM
        measurement
    WHERE
        date >= '{start_date}'
    """
start_df = pd.read_sql(query4, conn)
start_df
```

```python
# Fourth query to be inserted in the flask app.
# Return a JSON list of the minimum temperature, the average temperature, 
# and the max temperature for a given start-end range.

start_date = "2017-08-16"
end_date = "2017-08-31"

query5 = f"""
    SELECT
        min(tobs) as TMIN,
        avg(tobs) as TAVG,
        max(tobs) as TMAX
    FROM
        measurement
    WHERE
        date >= '{start_date}'
    AND 
        date <= '{end_date}'
    """
start_end_df = pd.read_sql(query5, conn)
start_end_df
```
